```
data.drop(['method'],axis=1,inplace=True)
```

```
data.info()
```

```
# Most Wins in Ipl by pandas
temp=pd.DataFrame({"Winner":data['winner']})
count_wins=temp.value_counts()
print(count_wins)

labels = [X[0]for X in count_wins.keys()]
bar,ax = plt.subplots(figsize=(12,8))
ax =plt.pie(x = count_wins,autopct='%.1f%%',labels= labels)
plt.title("Most wins in IPL", fontsize =17)
```

```
#Most wins in Eliminator
sns.countplot(data['winner'][data['eliminator']=='Y'],data=data)
plt.title('Most Wins in Eliminator')
plt.xticks(rotation=90)
```

```
 sns.barplot( y = data ['venue'].value_counts().head(8).values,
             x = data ['venue'].value_counts().head(8).index,
             data=data )
 plt.title('Most Famous')
 plt.xlabel('venue count')
 plt.ylabel('venue')
 plt.xticks(rotation=90)
```

```
 sns.barplot( x = data ['umpire1'].value_counts().head().index,
             y = data ['umpire1'].value_counts().head().values,
             data=data )
 plt.title('Most Umpire1')
 plt.ylabel('Match count')
 plt.xlabel('Top 5 Umpire')
 plt.xticks(rotation=45)
```

```
sns.barplot( x = data ['umpire2'].value_counts().head().index,
             y = data ['umpire2'].value_counts().head().values,
             data=data )
plt.title('Most Umpire2')
plt.ylabel('Match count')
plt.xlabel('Top 5 Umpire2')
plt.xticks(rotation=45)
 
 
```
